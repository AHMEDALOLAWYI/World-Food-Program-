# Resize images to 512x512

```python
# If True, will launch the resize
LAUNCH = True
```

```python
def center_crop(img, new_width=64, new_height=None):        

    width = img.width
    height = img.height


    if new_height is None:
        new_height = new_width

    left = int(np.ceil((width - new_width) / 2))
    right = width - int(np.floor((width - new_width) / 2))

    top = int(np.ceil((height - new_height) / 2))
    bottom = height - int(np.floor((height - new_height) / 2))
    
    #print(f'left {left}, right {right}, top {top}, bottom {bottom}')

   
    center_cropped_img = img.crop((left, top, right, bottom))
    
    assert((center_cropped_img.height == new_height) and (center_cropped_img.width == new_width))

    return center_cropped_img

def crop_on_img(filename, path_target, size=41):
    img = Image.open(filename)
    img = center_crop(img, size, size)
    print(path_target+filename.stem+'.png')
    img.save(path_target+filename.stem+'.png')
    #return img
```

```python
# Path where your downloaded data is
path_source = 'data/WFP_Nepal_RGB-Scale3.5_PNG/'
# Target path of resized data
path_target = 'data/task2-rgb/'


if LAUNCH:
    for r, d, f in walk(path_source):
        if r != path_source:
            break
        print(f'In directory {r}: {len(f)} files')
        for file in f:
            #print(f'working on img {path_source+file}')
            image = Path(path_source+file)    
            crop_on_img(image, path_target, 512)
```

# Add group id for each coordinate

Because there are several images per coordinate, we need to be careful not to  have some images of a coordinate in train set and other images of the same coordinate on val/test set. This would cause data leaks.

This section will add a group id (unique for a GPS entry) to the file name. This will then be easier to generate train / val set of different group ids.

```python
import os 
```

```python
def get_coord_from_name(file_name):
    #print(file_name)
    x = file_name.split('_')[2]
    y = file_name.split('_')[3]
    
#     x = x[:2]+'.'+x[3:]
#     y = y[:2]+'.'+y[3:]

    x = x.replace('-', '.')
    y = y.replace('-', '.')
     
    return np.float64(x), np.float64(y)
    
```

```python
img_ref = {}
img_nb = {}
group_index = 0
for r, d, f in walk(path_target):
    if r != path_target:
        break
    print(f'In directory {r}: {len(f)} files')
    for file in f:
        x, y = get_coord_from_name(file)
        assert((x > 84) and (y > 27))
        key = str(x)+'_' +str(y)
        
        if not key in img_ref:
            print('new image')
            img_ref[key] = group_index
            img_nb[key] = 0
            
            
            os.rename(path_target+file, path_target+str(group_index)+'-'+file)
            group_index += 1
            
        else:
            img_nb[key] = img_nb[key] + 1
            gr_im = img_ref[key]
            os.rename(path_target+file, path_target+str(gr_im)+'-'+file)
        
            
```
